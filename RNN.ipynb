```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("neisse/scrapped-lyrics-from-6-genres")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
import os

print(os.listdir(path))
```

```python
df = pd.read_csv(
    os.path.join(path, "lyrics-data.csv")
)
```

```python
df.head()
```

```python
df = df.dropna()

df = df.head(200)

text = " ".join(df["Lyric"].astype(str))

print("Total characters:", len(text))
```

```python
import re

text = text.lower()

text = re.sub(r'[^a-zA-Z\s]', '', text)

text = re.sub(r'\s+', ' ', text)

print(text[:500])
```

```python
tokenizer = Tokenizer(num_words=5000)

tokenizer.fit_on_texts([text])

total_words = 5000

print("Vocabulary Size:", total_words)
```

```python
words = text.split()

sequences = []

window_size = 5

for i in range(window_size, len(words)):

    seq = words[i-window_size:i+1]

    sequences.append(seq)

print("Total sequences:", len(sequences))
```

```python
input_sequences = []

for seq in sequences:

    encoded = tokenizer.texts_to_sequences([" ".join(seq)])[0]

    if len(encoded) == window_size + 1:
        input_sequences.append(encoded)

input_sequences = np.array(input_sequences)

print(input_sequences.shape)
```

```python
X = input_sequences[:, :-1]

y = input_sequences[:, -1]

y = to_categorical(
    y,
    num_classes=total_words
)

print(X.shape)
print(y.shape)
```

```python
model = Sequential([

    Embedding(
        input_dim=total_words,
        output_dim=64,
        input_length=window_size
    ),

    SimpleRNN(128),

    Dense(
        total_words,
        activation='softmax'
    )

])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    X,
    y,
    epochs=50,
    batch_size=128,
    validation_split=0.2
)
```

```python
def predict_next_word(seed_text):

    token_list = tokenizer.texts_to_sequences(
        [seed_text]
    )[0]

    token_list = token_list[-window_size:]

    token_list = pad_sequences(
        [token_list],
        maxlen=window_size
    )

    predicted = np.argmax(
        model.predict(token_list, verbose=0)
    )

    for word, index in tokenizer.word_index.items():

        if index == predicted:

            return word

    return ""
```

```python
def generate_text(seed_text, n_words):

    for _ in range(n_words):

        next_word = predict_next_word(seed_text)

        seed_text += " " + next_word

    return seed_text
```

```python
print(generate_text("we will we will", 5))

print(generate_text("hello from the", 5))

print(generate_text("i love you", 5))
```
